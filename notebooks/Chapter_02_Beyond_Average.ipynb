## Part 2: Simulating Normal Variation

```python
# Simulate what happens when we flip a fair coin 50 times, repeated 10,000 times
n_simulations = 10000
n_flips = 50

# Run simulations
simulation_results = [np.sum(np.random.binomial(1, 0.5, n_flips)) 
                      for _ in range(n_simulations)]

# Create histogram
fig, ax = plt.subplots(figsize=(14, 8))

# Plot histogram
counts, bins, patches = ax.hist(simulation_results, bins=range(10, 41), 
                                 alpha=0.7, color='steelblue', edgecolor='black')

# Mark Kabir's and Ananya's results
ax.axvline(x=kabir_heads, color='red', linestyle='--', linewidth=3, 
           label=f"Kabir's result ({kabir_heads} heads)", alpha=0.8)
ax.axvline(x=ananya_heads, color='orange', linestyle='--', linewidth=3, 
           label=f"Ananya's result ({ananya_heads} heads)", alpha=0.8)
ax.axvline(x=25, color='green', linestyle='-', linewidth=3, 
           label='Expected (25 heads)', alpha=0.8)

# Labels and title
ax.set_xlabel('Number of Heads (out of 50 flips)', fontsize=13, fontweight='bold')
ax.set_ylabel('Frequency (out of 10,000 simulations)', fontsize=13, fontweight='bold')
ax.set_title('What Does "Normal Variation" Look Like?\n' +
             '10,000 simulations of flipping a fair coin 50 times',
             fontsize=16, fontweight='bold', pad=20)
ax.legend(fontsize=12, loc='upper left')
ax.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()

# Calculate statistics
percent_at_24 = sum(1 for x in simulation_results if x == 24) / n_simulations * 100
percent_at_27 = sum(1 for x in simulation_results if x == 27) / n_simulations * 100
percent_at_or_more_extreme_than_27 = sum(1 for x in simulation_results if x >= 27 or x <= 23) / n_simulations * 100

print("\nðŸ“Š ANALYSIS:\n")
print(f"Peak of the bell curve: {np.argmax(counts) + 10} heads")
print(f"\nHow common are results like theirs?")
print(f"  Exactly 24 heads (Ananya): {percent_at_24:.1f}% of simulations")
print(f"  Exactly 27 heads (Kabir): {percent_at_27:.1f}% of simulations")
print(f"  27 or more extreme: {percent_at_or_more_extreme_than_27:.1f}% of simulations")
print("\nâœ… CONCLUSION: Both results are well within normal variation!")
```

## FIGURE 2.3: Visual Metaphor â€“ Target Practice

```python
# Create target practice visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))

# Function to draw a target
def draw_target(ax, title, tight_cluster=True):
    # Draw concentric circles
    circles = [10, 7, 5, 3, 1]
    colors = ['white', 'lightcoral', 'white', 'lightcoral', 'red']
    for radius, color in zip(circles, colors):
        circle = plt.Circle((0, 0), radius, color=color, ec='black', linewidth=2)
        ax.add_patch(circle)
    
    # Add dots representing shots
    np.random.seed(42)
    if tight_cluster:
        # Player A: tight cluster near center
        dots_x = np.random.normal(0, 1.2, 20)
        dots_y = np.random.normal(0, 1.2, 20)
    else:
        # Player B: scattered wildly
        dots_x = np.random.normal(0, 4.5, 20)
        dots_y = np.random.normal(0, 4.5, 20)
    
    ax.scatter(dots_x, dots_y, s=150, c='darkblue', marker='o', 
               edgecolors='black', linewidth=1.5, alpha=0.7, zorder=10)
    
    # Styling
    ax.set_xlim([-12, 12])
    ax.set_ylim([-12, 12])
    ax.set_aspect('equal')
    ax.axis('off')
    ax.set_title(title, fontsize=14, fontweight='bold', pad=20)

# Draw both targets
draw_target(ax1, 'Player A: Precise and Predictable\n"The Marksman"', tight_cluster=True)
draw_target(ax2, 'Player B: Chaotic and Unpredictable\n"The Gambler"', tight_cluster=False)

fig.suptitle('FIGURE 2.3: Imagine the Scores as Balls Thrown at a Target\n' +
             'Tight cluster = consistent. Wide scatter = unpredictable.',
             fontsize=16, fontweight='bold', y=0.98)

plt.tight_layout()
plt.show()

print("\nðŸŽ¯ METAPHOR:")
print("Player A hits the same spot every time â€” boring in the best way")
print("Player B is all over the place â€” occasionally brilliant, often disastrous")
print("\nBoth might average the same distance from center,")
print("but you'd trust them for different tasks!")
```

### Variance and Standard Deviation

"Smart question. You're rightâ€”range ignores most of the data. That's why statisticians invented something called **variance**, and its cousin, **standard deviation**." He paused. "Do you want the math, or just the concept?"

"Concept first," Kabir said. "Then maybe the math if it's not too scary."

"Fair. The concept is simple: **variance measures the typical distance of each score from the average.** If Player A's scores are all close to thirty-three, variance is low. If Player B's scores are far from thirty-three, variance is high."

"How do you calculate it?" Ananya asked, pen poised.

"For each score, you calculate how far it is from the average. Then you square those distancesâ€”so negative and positive differences don't cancel out. Then you average all those squared distances. That's variance."

**Variance = Average of (each score - mean score)Â²**

"And standard deviation is just the square root of variance. It brings you back to the original unitsâ€”runs in this case, instead of runs-squared."

```python
# Demonstrate variance calculation step-by-step with a simple example
simple_data = [3, 5, 5, 7]
mean_simple = np.mean(simple_data)

print("\nðŸ”¢ CALCULATING VARIANCE: STEP BY STEP\n")
print(f"Simple dataset: {simple_data}\n")

print(f"Step 1: Calculate mean")
print(f"  Mean = ({' + '.join(map(str, simple_data))}) / {len(simple_data)} = {mean_simple}\n")

print(f"Step 2: Find distance from mean for each value")
distances = [x - mean_simple for x in simple_data]
for i, (val, dist) in enumerate(zip(simple_data, distances)):
    print(f"  {val} - {mean_simple} = {dist:+.1f}")

print(f"\nStep 3: Square those distances (so negatives don't cancel)")
squared_distances = [d**2 for d in distances]
for i, (dist, sq) in enumerate(zip(distances, squared_distances)):
    print(f"  ({dist:+.1f})Â² = {sq:.1f}")

print(f"\nStep 4: Average the squared distances = Variance")
variance_simple = np.mean(squared_distances)
print(f"  Variance = ({' + '.join(map(lambda x: f'{x:.1f}', squared_distances))}) / {len(squared_distances)}")
print(f"  Variance = {variance_simple:.2f}\n")

print(f"Step 5: Take square root = Standard Deviation")
std_simple = np.sqrt(variance_simple)
print(f"  Standard Deviation = âˆš{variance_simple:.2f} = {std_simple:.2f}")

print(f"\nâœ… This data has mean {mean_simple} and standard deviation {std_simple:.2f}")
print(f"Most points are within about {std_simple:.2f} units of the center.")
```

```python
# Now apply to our cricket players
print("\nðŸ APPLYING TO OUR CRICKET PLAYERS:\n")

print(f"Player A:")
print(f"  Average: {avg_A:.1f} runs")
print(f"  Standard Deviation: {std_A:.1f} runs")
print(f"  â†’ Typically varies by about {std_A:.1f} runs from the average\n")

print(f"Player B:")
print(f"  Average: {avg_B:.1f} runs")
print(f"  Standard Deviation: {std_B:.1f} runs")
print(f"  â†’ Typically varies by about {std_B:.1f} runs from the average\n")

print("ðŸ’¡ HUGE DIFFERENCE!")
print(f"Player B's standard deviation is {std_B/std_A:.1f}x larger than Player A's!")
```

```python
# Example: Tracking study hours for a week
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
study_hours = [2.5, 3.0, 1.5, 4.0, 2.0, 5.5, 3.5]  # Your data here!

# Calculate statistics
mean_study = np.mean(study_hours)
std_study = np.std(study_hours, ddof=1)
range_study = max(study_hours) - min(study_hours)

print("\nðŸ“š MY STUDY HOURS ANALYSIS\n")
print("Daily hours:")
for day, hours in zip(days, study_hours):
    print(f"  {day}: {hours} hours")

print(f"\nStatistics:")
print(f"  Average: {mean_study:.1f} hours/day")
print(f"  Range: {range_study:.1f} hours")
print(f"  Standard Deviation: {std_study:.1f} hours")

print(f"\nðŸ’¡ Interpretation:")
if std_study < 1.0:
    print("You're quite consistent in your study habits!")
elif std_study < 1.5:
    print("Moderate variation - some days more focused than others")
else:
    print("High variability - your study schedule is quite erratic!")

# Visualize
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(days, study_hours, color='skyblue', edgecolor='black', alpha=0.7)
ax.axhline(y=mean_study, color='red', linestyle='--', linewidth=2, 
           label=f'Average: {mean_study:.1f} hours', alpha=0.7)
ax.set_ylabel('Hours', fontsize=12, fontweight='bold')
ax.set_title('My Study Hours This Week', fontsize=14, fontweight='bold')
ax.legend(fontsize=11)
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()
```

## ðŸ“š Key Concepts from Chapter 2

### 1. Law of Large Numbers (Intuitive)
- Small samples (like 50 coin flips) can show lots of variation
- With more data, the proportion stabilizes near the true value
- Be suspicious of conclusions drawn from tiny samples

### 2. Normal Variation vs. Genuine Anomaly
- Random processes naturally vary
- Not every deviation from expected is meaningful
- Learning to distinguish signal from noise is crucial

### 3. Average Isn't Everything
- Two datasets can have the same average but tell completely different stories
- Need to know both center (average) and spread (variation)
- **The distribution shape matters as much as the central tendency**

### 4. Measures of Spread
- **Range**: Simplest measure (max - min), but only uses two data points
- **Variance**: Average squared distance from the mean
- **Standard Deviation**: Square root of variance, easier to interpret

### 5. Context Determines Value
- Sometimes you want consistency (Player A, steady farmer)
- Sometimes you want high variance (Player B when desperate, venture capital)
- Understanding your goals helps you interpret variation
