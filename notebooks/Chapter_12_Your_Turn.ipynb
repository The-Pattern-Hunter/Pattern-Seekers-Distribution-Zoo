## Part 3: The Eight Habits of Mind

### Professor Mishra's Framework

At their one-year anniversary meeting, Professor Mishra named what the students had developed:

*"You've developed a scientific mindset. Not scientific knowledgeâ€”that comes and goes, updates with new discoveries. A scientific **mindset**â€”that stays."*

He wrote on his whiteboard:

### **THE PATTERN SEEKER'S HABITS OF MIND**

```python
# Visualization: The Eight Habits of Mind

fig, ax = plt.subplots(figsize=(14, 10))
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.axis('off')

# Title
ax.text(5, 9.5, 'THE PATTERN SEEKER\'S HABITS OF MIND', 
        ha='center', va='top', fontsize=18, fontweight='bold',
        bbox=dict(boxstyle='round,pad=0.8', facecolor='gold', 
                 edgecolor='darkgoldenrod', linewidth=3))

# Eight habits arranged in a circular pattern
habits = [
    ("1. Pattern Recognition", "See distributions everywhere\nStructure in chaos"),
    ("2. Probabilistic Thinking", "Think in likelihoods\nNot certainties"),
    ("3. Model-Based Reasoning", "Build simplified versions\nHold models lightly"),
    ("4. Healthy Skepticism", "Question claims\nCheck sources"),
    ("5. Quantification", "Measure instead of guess\nData over feeling"),
    ("6. Visualization as Thinking", "Graphs discover patterns\nNot just show results"),
    ("7. Iterative Refinement", "First model will be wrong\nTest, learn, improve"),
    ("8. Comfort with Uncertainty", "Don't need all answers\nProbability is enough")
]

# Arrange in circular pattern
center_x, center_y = 5, 5
radius = 3.5
angles = np.linspace(0, 2*np.pi, len(habits), endpoint=False) + np.pi/2

colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A', 
          '#98D8C8', '#F7DC6F', '#BB8FCE', '#85C1E2']

for i, (habit_title, habit_desc) in enumerate(habits):
    x = center_x + radius * np.cos(angles[i])
    y = center_y + radius * np.sin(angles[i])
    
    # Draw box for each habit
    box = FancyBboxPatch((x-0.8, y-0.5), 1.6, 1.0, 
                          boxstyle="round,pad=0.1", 
                          facecolor=colors[i], 
                          edgecolor='black', 
                          linewidth=2, 
                          alpha=0.8)
    ax.add_patch(box)
    
    # Add text
    ax.text(x, y+0.2, habit_title, 
            ha='center', va='center', 
            fontsize=9, fontweight='bold', 
            color='black')
    ax.text(x, y-0.2, habit_desc, 
            ha='center', va='center', 
            fontsize=7, color='black',
            style='italic')
    
    # Draw line from center to habit
    ax.plot([center_x, x], [center_y, y], 'k--', linewidth=1, alpha=0.3)

# Central core
core = Circle((center_x, center_y), 0.8, 
              facecolor='white', 
              edgecolor='darkgoldenrod', 
              linewidth=4)
ax.add_patch(core)
ax.text(center_x, center_y, 'PATTERN\nSEEKER', 
        ha='center', va='center', 
        fontsize=12, fontweight='bold',
        color='darkgoldenrod')

# Add inspirational quote at bottom
ax.text(5, 0.5, '"These aren\'t just math skillsâ€”they\'re life skills." - Professor Mishra', 
        ha='center', va='center', fontsize=11, style='italic',
        bbox=dict(boxstyle='round,pad=0.5', facecolor='lightyellow', alpha=0.8))

plt.title('Eight Habits That Change How You Think', 
          fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()

print("\nðŸŽ“ The Eight Habits Explained:")
print("\n1. PATTERN RECOGNITION - You see distributions everywhere now")
print("2. PROBABILISTIC THINKING - You think in likelihoods, not certainties")
print("3. MODEL-BASED REASONING - You build simplified versions of reality")
print("4. HEALTHY SKEPTICISM - You question claims and check sources")
print("5. QUANTIFICATION - You measure instead of guessing")
print("6. VISUALIZATION AS THINKING - Graphs help you discover, not just show")
print("7. ITERATIVE REFINEMENT - You expect first models to be wrong")
print("8. COMFORT WITH UNCERTAINTY - Probability is enough")
print("\nðŸ’¡ Together, these habits form a SCIENTIFIC MINDSET")
```

```python
# Final Visualization: The Learning Spiral

fig, ax = plt.subplots(figsize=(12, 10))
ax.set_xlim(-6, 6)
ax.set_ylim(-1, 11)
ax.axis('off')

# Create upward spiral
theta = np.linspace(0, 6*np.pi, 1000)
r = theta / (2*np.pi)  # Radius increases with angle
x = r * np.cos(theta)
y = r * np.sin(theta) / 2 + theta / (2*np.pi) * 1.5  # Upward component

# Plot spiral with gradient color
for i in range(len(x)-1):
    color = plt.cm.RdYlGn(i / len(x))  # Gradient from red to yellow to green
    ax.plot(x[i:i+2], y[i:i+2], color=color, linewidth=3)

# Add labels at different points in spiral
stages = [
    (0, 0, 'Curiosity', 0),
    (0.8, 0.7, 'Question', np.pi/2),
    (-0.3, 1.4, 'Collect Data', np.pi),
    (-1.2, 2.1, 'Build Model', 3*np.pi/2),
    (0.5, 2.8, 'Test', 2*np.pi),
    (1.5, 3.5, 'Learn', 5*np.pi/2),
    (0, 4.2, 'New Questions', 3*np.pi),
    (-1.8, 4.9, 'Better Model', 7*np.pi/2),
    (1.5, 5.6, 'Deeper\nUnderstanding', 4*np.pi),
]

for x_pos, y_pos, label, angle in stages:
    # Add point
    ax.scatter(x_pos, y_pos, s=200, c='gold', edgecolors='black', linewidths=2, zorder=5)
    # Add label
    ax.text(x_pos, y_pos, label, ha='center', va='center', 
            fontsize=10, fontweight='bold', zorder=6)

# Add arrows to show direction
arrow_positions = [(2, 3), (4, 5.5), (5, 7.5)]
for i, (idx1, idx2) in enumerate(arrow_positions):
    x1, y1 = x[int(idx1*len(x)/6)], y[int(idx1*len(y)/6)]
    x2, y2 = x[int(idx2*len(x)/6)], y[int(idx2*len(y)/6)]
    ax.annotate('', xy=(x2, y2), xytext=(x1, y1),
                arrowprops=dict(arrowstyle='->', lw=3, color='darkgoldenrod'))

# Title and notes
ax.text(0, 10, 'THE LEARNING SPIRAL', ha='center', va='top', 
        fontsize=18, fontweight='bold',
        bbox=dict(boxstyle='round,pad=0.8', facecolor='gold', 
                 edgecolor='darkgoldenrod', linewidth=3))

ax.text(0, -0.5, 'Learning is not linearâ€”it\'s a spiral.\nEach loop brings deeper understanding, new questions, better models.\nThe journey never truly ends.', 
        ha='center', va='top', fontsize=11, style='italic',
        bbox=dict(boxstyle='round,pad=0.5', facecolor='lightyellow', alpha=0.8))

# Add progression indicators
ax.text(-5, 0, 'START\n(Confusion)', ha='center', va='center', 
        fontsize=10, color='darkred', fontweight='bold',
        bbox=dict(boxstyle='round', facecolor='lightcoral', alpha=0.7))

ax.text(5, 8, 'MASTERY\n(Never Complete)', ha='center', va='center', 
        fontsize=10, color='darkgreen', fontweight='bold',
        bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.7))

plt.title('"It\'s a spiralâ€”each loop brings deeper understanding." - Professor Mishra', 
          fontsize=14, style='italic', pad=20)
plt.tight_layout()
plt.show()

print("\nðŸŒ€ The Learning Spiral:")
print("\nScience isn't linear. Learning isn't linear. It's a spiral:")
print("  â€¢ Each loop you return to similar questions")
print("  â€¢ But each time, you understand more deeply")
print("  â€¢ Each iteration: Curiosity â†’ Question â†’ Data â†’ Model â†’ Test â†’ Learn")
print("  â€¢ Then back to curiosity... but at a higher level")
print("\nðŸ’¡ You never 'finish' learning. You just keep spiraling upward.")
print("\nðŸŽ“ Ananya, Kabir, and Priya are one year into their spiral.")
print("   Where are you in yours?")
```
