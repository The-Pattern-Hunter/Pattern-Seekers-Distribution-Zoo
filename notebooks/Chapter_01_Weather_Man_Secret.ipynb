## üìä PART 2: The Mystery Data

Let's load the actual rainfall data that Ananya used ‚Äî 50 years of monsoon records from Sambalpur district, Odisha!

**In the book:** Professor Mishra shows Ananya and Kabir a scatter plot of 50 years of rainfall. Let's recreate that exact moment.

```python
# Create the historical rainfall data (in millimeters)
# This represents 50 years of monsoon season totals in Sambalpur

# Each number is one year's total rainfall during monsoon (June-September)
rainfall_data = np.array([
    410, 389, 425, 398, 456, 401, 387, 432, 394, 418,  # 1974-1983
    402, 445, 378, 412, 396, 428, 408, 391, 434, 405,  # 1984-1993
    419, 393, 441, 415, 382, 427, 399, 448, 412, 386,  # 1994-2003
    423, 407, 392, 437, 414, 395, 429, 403, 416, 388,  # 2004-2013
    442, 409, 397, 433, 411, 385, 424, 406, 420, 389   # 2014-2023
])

# Create years array (1974 to 2023)
years = np.arange(1974, 2024)

print(f"üìö Loaded {len(rainfall_data)} years of rainfall data")
print(f"üìÖ From {years[0]} to {years[-1]}")
print(f"")
print(f"üîç Quick peek at first 10 years:")
print(f"   Years: {years[:10].tolist()}")
print(f"   Rainfall (mm): {rainfall_data[:10].tolist()}")
```

## üìà PART 3: Figure 1.1 - The Scatter Plot

**Recreation of Figure 1.1 from the book:** *"Fifty Years of Monsoon - Each dot is one season's story"*

üëâ **Run this cell to see the pattern (or chaos?):**

```python
# Create the scatter plot
plt.figure(figsize=(12, 6))

# Plot each year as a dot
plt.scatter(years, rainfall_data,
           color='steelblue',      # Color of the dots
           s=80,                   # Size of dots
           alpha=0.6,              # Transparency (0=invisible, 1=solid)
           edgecolors='navy',      # Border color
           linewidth=1.5)

# Add labels and title
plt.xlabel('Year', fontsize=12, fontweight='bold')
plt.ylabel('Total Monsoon Rainfall (mm)', fontsize=12, fontweight='bold')
plt.title('Figure 1.1: Fifty Years of Monsoon\nEach dot is one season\'s story',
         fontsize=14, fontweight='bold', pad=20)

# Add a horizontal line showing the average
mean_rainfall = np.mean(rainfall_data)
plt.axhline(y=mean_rainfall,
           color='red',
           linestyle='--',
           linewidth=2,
           label=f'Average: {mean_rainfall:.1f}mm')

plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\nü§î Look closely at the scatter plot:")
print(f"   ‚Ä¢ Average rainfall: {mean_rainfall:.1f}mm")
print(f"   ‚Ä¢ Highest year: {np.max(rainfall_data)}mm")
print(f"   ‚Ä¢ Lowest year: {np.min(rainfall_data)}mm")
print(f"   ‚Ä¢ Range: {np.max(rainfall_data) - np.min(rainfall_data)}mm difference!")
```

## üî¨ PART 4: The Histogram - Where the Magic Happens

### What's a Histogram?

Instead of showing *when* each rainfall happened (scatter plot), let's show *how often* different rainfall amounts occur.

**Think of it like:**
- Sorting your exam scores into buckets: 0-10, 10-20, 20-30, etc.
- Then counting how many scores fall in each bucket
- The taller the bar, the more common that range is!

**From the book:** This is where Ananya sees the bell curve for the first time...

```python
# Create Figure 1.2: The Histogram
plt.figure(figsize=(12, 7))

# Create the histogram
# bins=12 means we divide the data into 12 buckets
counts, bins, patches = plt.hist(rainfall_data,
                                bins=12,
                                color='skyblue',
                                edgecolor='navy',
                                alpha=0.7,
                                linewidth=1.5)

# Add labels
plt.xlabel('Rainfall Amount (mm)', fontsize=12, fontweight='bold')
plt.ylabel('Number of Years', fontsize=12, fontweight='bold')
plt.title('Figure 1.2: The Histogram Reveals the Pattern\n"It\'s not random‚Äîit has a shape!"',
         fontsize=14, fontweight='bold', pad=20)

# Add the mean line
plt.axvline(x=mean_rainfall,
           color='red',
           linestyle='--',
           linewidth=2.5,
           label=f'Average: {mean_rainfall:.1f}mm')

plt.legend(fontsize=11)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print("\n‚ú® THE PATTERN EMERGES!")
print("\nüìä What do you notice?")
print("   ‚Ä¢ The bars form a hill shape")
print("   ‚Ä¢ Most years cluster around the middle (400-420mm)")
print("   ‚Ä¢ Extreme values (very high or very low) are rare")
print("   ‚Ä¢ It's roughly symmetric - left side mirrors right side")
print("\nüéØ This shape is called a BELL CURVE (or Normal Distribution)")
print("   It's nature's favorite pattern!")
```

### üßí ELI5: Reading a Histogram

Imagine you asked 50 classmates their height and made piles of coins:
- One pile for "140-150 cm" students
- One pile for "150-160 cm" students
- One pile for "160-170 cm" students
- etc.

The histogram shows those piles!
- **Horizontal axis (x):** The ranges (buckets)
- **Vertical axis (y):** How many in each pile (frequency)
- **Tall bars:** Common values (lots of students)
- **Short bars:** Rare values (few students)

**For rainfall:**
- The tallest bar is around 405-420mm
- That means: "Most years had rainfall in this range"
- The short bars at edges mean: "Extreme rainfall is rare"

## üîî PART 5: The Bell Curve - Nature's Favorite Shape

**From the book:**
> *"The bell curve," Professor Mishra said. "Also called the normal distribution. It appears everywhere in nature. Heights, weights, test scores, measurement errors, rainfall..."*

Let's overlay a smooth bell curve on our histogram to see how well the pattern fits!

```python
# Calculate statistics for the bell curve
mean = np.mean(rainfall_data)
std_dev = np.std(rainfall_data)

print(f"üìê Rainfall Statistics:")
print(f"   Mean (Œº): {mean:.2f}mm")
print(f"   Standard Deviation (œÉ): {std_dev:.2f}mm")
print(f"")
print(f"üéØ What this means:")
print(f"   ‚Ä¢ Center of the bell curve: {mean:.1f}mm")
print(f"   ‚Ä¢ About 68% of years fall within {mean-std_dev:.1f} to {mean+std_dev:.1f}mm")
print(f"   ‚Ä¢ About 95% of years fall within {mean-2*std_dev:.1f} to {mean+2*std_dev:.1f}mm")
```

```python
# Create Figure 1.3: Histogram with Bell Curve Overlay
from scipy import stats

plt.figure(figsize=(12, 7))

# Histogram (density=True makes it comparable to the curve)
plt.hist(rainfall_data,
        bins=12,
        density=True,
        color='skyblue',
        edgecolor='navy',
        alpha=0.7,
        linewidth=1.5,
        label='Actual Data')

# Generate smooth bell curve
x = np.linspace(rainfall_data.min() - 20, rainfall_data.max() + 20, 100)
bell_curve = stats.norm.pdf(x, mean, std_dev)

# Plot the bell curve
plt.plot(x, bell_curve,
        color='red',
        linewidth=3,
        label='Perfect Bell Curve')

# Mark the mean
plt.axvline(x=mean,
           color='darkred',
           linestyle='--',
           linewidth=2,
           label=f'Mean: {mean:.1f}mm')

# Shade the ¬±1 standard deviation region (68% of data)
plt.axvspan(mean - std_dev, mean + std_dev,
           alpha=0.2,
           color='yellow',
           label='¬±1 SD (68% of years)')

plt.xlabel('Rainfall Amount (mm)', fontsize=12, fontweight='bold')
plt.ylabel('Probability Density', fontsize=12, fontweight='bold')
plt.title('Figure 1.3: The Bell Curve Emerges\n"Chaos has a shape ‚Äî and it\'s beautiful!"',
         fontsize=14, fontweight='bold', pad=20)
plt.legend(fontsize=10, loc='upper left')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print("\nüéâ THE BIG DISCOVERY:")
print("\n   Real rainfall data follows a bell curve!")
print("   ‚Ä¢ It's not perfectly random")
print("   ‚Ä¢ It's not perfectly predictable")
print("   ‚Ä¢ It's PATTERNED UNCERTAINTY")
print("\nüîÆ This means we can make PROBABILISTIC predictions:")
print(f"   'Next year\'s rainfall will PROBABLY be between")
print(f"    {mean-std_dev:.0f}mm and {mean+std_dev:.0f}mm (68% confident)'")
```

### üéØ EXPERIMENT 2: What if rainfall became more erratic?

**Challenge:** Make the rainfall more variable (more spread out) while keeping the same average.

```python
# Generate more variable rainfall (same mean, bigger standard deviation)
mean_rainfall = np.mean(rainfall_data)
original_std = np.std(rainfall_data)

# Generate new data: same mean, but 2x the spread
erratic_rainfall = np.random.normal(mean_rainfall, original_std * 2, len(rainfall_data))

# Plot comparison
plt.figure(figsize=(12, 6))

plt.hist(rainfall_data, bins=12, alpha=0.5, color='green',
        label=f'Original (SD={original_std:.1f})', edgecolor='darkgreen')
plt.hist(erratic_rainfall, bins=12, alpha=0.5, color='orange',
        label=f'Erratic (SD={np.std(erratic_rainfall):.1f})', edgecolor='darkorange')

plt.axvline(mean_rainfall, color='red', linestyle='--', linewidth=2, label='Same Mean')

plt.xlabel('Rainfall (mm)', fontweight='bold')
plt.ylabel('Number of Years', fontweight='bold')
plt.title('Experiment: What If Rainfall Became More Erratic?', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print(f"\nüìä Results:")
print(f"   Both datasets have same mean: {mean_rainfall:.1f}mm")
print(f"   Original spread (SD): {original_std:.1f}mm")
print(f"   Erratic spread (SD): {np.std(erratic_rainfall):.1f}mm")
print(f"\nüí° Notice: Same center, different PREDICTABILITY!")
print(f"   Green = predictable (tight cluster)")
print(f"   Orange = unpredictable (wide spread)")
print(f"\nüåæ For farmers: Which would you prefer?")
print(f"   Steady rain (green) or unpredictable rain (orange)?")
```

## üì± PART 7: The Modern Hook - Delivery Time Data

**From the book:** Kabir was tracking his food delivery app. Let's analyze that data too!

Remember: The app predicted 28 minutes, then 32, then 36... Why do predictions keep changing?

```python
# Kabir's food delivery times (in minutes) over 30 orders
delivery_times = np.array([
    28, 32, 25, 35, 29, 31, 27, 33, 30, 28,
    34, 26, 31, 29, 32, 28, 30, 27, 33, 29,
    31, 28, 35, 30, 32, 29, 31, 28, 30, 33
])

# Calculate statistics
avg_delivery = np.mean(delivery_times)
std_delivery = np.std(delivery_times)

print(f"üçú Kabir's Delivery Data Analysis:")
print(f"   Average delivery time: {avg_delivery:.1f} minutes")
print(f"   Standard deviation: {std_delivery:.1f} minutes")
print(f"   Fastest delivery: {np.min(delivery_times)} minutes")
print(f"   Slowest delivery: {np.max(delivery_times)} minutes")
print(f"\nüì± App's dilemma: What should it predict?")
print(f"   If it says {avg_delivery:.0f} minutes, it's right on average...")
print(f"   But actual time varies by ¬±{std_delivery:.0f} minutes!")
```

```python
# Visualize delivery times
plt.figure(figsize=(12, 6))

plt.hist(delivery_times, bins=8, color='coral', edgecolor='darkred', alpha=0.7)
plt.axvline(avg_delivery, color='blue', linestyle='--', linewidth=2.5,
           label=f'Average: {avg_delivery:.1f} min')

# Shade the typical range (¬±1 SD)
plt.axvspan(avg_delivery - std_delivery, avg_delivery + std_delivery,
           alpha=0.2, color='yellow', label='¬±1 SD (typical range)')

plt.xlabel('Delivery Time (minutes)', fontweight='bold')
plt.ylabel('Number of Orders', fontweight='bold')
plt.title('Figure 1.5: Kabir\'s Delivery Time Distribution\n"Even momos follow a bell curve!"',
         fontsize=14, fontweight='bold', pad=20)
plt.legend(fontsize=11)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print("\n‚ú® PATTERN SPOTTED AGAIN!")
print("   ‚Ä¢ Bell curve shape (even for delivery times!)")
print("   ‚Ä¢ Most deliveries around 28-32 minutes")
print("   ‚Ä¢ Rare to see 25min (super fast) or 35min (delayed)")
print("\nüéØ The app can't predict EXACTLY when food arrives...")
print("   But it CAN predict a RANGE with confidence!")
print(f"   \"Your food will arrive in {avg_delivery-std_delivery:.0f}-{avg_delivery+std_delivery:.0f} minutes (68% sure)\"")
```

```python
# üéØ YOUR DATA HERE:
# Replace these sample numbers with YOUR real data!

my_data = [7, 6.5, 8, 7, 6, 8.5, 7, 7.5, 6.5, 8, 7, 6.5, 7.5, 8]  # Example: hours of sleep
data_name = "My Sleep Hours"  # Give your data a name!
data_units = "hours"  # What are you measuring?

# Convert to numpy array
my_data = np.array(my_data)

# Calculate statistics
my_mean = np.mean(my_data)
my_std = np.std(my_data)
my_min = np.min(my_data)
my_max = np.max(my_data)

# Create visualization
plt.figure(figsize=(12, 7))

# Histogram
plt.hist(my_data, bins=min(8, len(my_data)//2),
        color='mediumseagreen', edgecolor='darkgreen', alpha=0.7, linewidth=1.5)

# Add mean line
plt.axvline(my_mean, color='red', linestyle='--', linewidth=2.5,
           label=f'Average: {my_mean:.2f} {data_units}')

# Shade typical range
plt.axvspan(my_mean - my_std, my_mean + my_std,
           alpha=0.2, color='yellow', label='Typical range (¬±1 SD)')

plt.xlabel(f'{data_name} ({data_units})', fontweight='bold', fontsize=12)
plt.ylabel('Frequency', fontweight='bold', fontsize=12)
plt.title(f'My Pattern Discovery: {data_name}', fontsize=14, fontweight='bold', pad=20)
plt.legend(fontsize=11)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Print analysis
print(f"\nüìä YOUR DATA ANALYSIS:")
print(f"\n   Dataset: {data_name}")
print(f"   Number of data points: {len(my_data)}")
print(f"   Average: {my_mean:.2f} {data_units}")
print(f"   Standard deviation: {my_std:.2f} {data_units}")
print(f"   Range: {my_min:.2f} to {my_max:.2f} {data_units}")
print(f"\nüéØ PATTERN INSIGHTS:")
print(f"   ‚Ä¢ Your typical {data_name.lower()}: {my_mean:.1f} {data_units}")
print(f"   ‚Ä¢ Your usual range: {my_mean-my_std:.1f} to {my_mean+my_std:.1f} {data_units}")
print(f"   ‚Ä¢ Variability: {'High' if my_std/my_mean > 0.2 else 'Moderate' if my_std/my_mean > 0.1 else 'Low'}")

# Personalized insight
if my_std/my_mean < 0.1:
    print(f"\nüí° You're very CONSISTENT! Your {data_name.lower()} doesn't vary much.")
elif my_std/my_mean < 0.2:
    print(f"\nüí° You have MODERATE variability. Some days differ, but there's a pattern.")
else:
    print(f"\nüí° You have HIGH variability! Your {data_name.lower()} changes a lot day-to-day.")

print(f"\nüåü Congratulations! You just did real statistical analysis!")
```
